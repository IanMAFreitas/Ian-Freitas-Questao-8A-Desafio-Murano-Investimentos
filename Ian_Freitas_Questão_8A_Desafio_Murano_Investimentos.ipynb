```python
import matplotlib.pyplot as plt

def Q8_A(stock):
  import matplotlib.pyplot as plt
  import pandas as pd

  stock_ = stock + '.us.txt'
  df = pd.read_csv(stock_, index_col='Date')['Close']
  
  df.plot(figsize=(10,70), label=stock.upper())
  df.rolling(10).mean().plot(label='MM10')
  df.rolling(20).mean().plot(label='MM20')
  df.rolling(50).mean().plot(label='MM50')
  df.rolling(200).mean().plot(label='MM200')

  plt.legend()
  plt.title(f'{stock.upper()} Medias Moveis')
  
stocks_list = ['aapl','elec','fss','ghy','googl','mgu','mjco','msft','ndsn','shak']

for count, stock in enumerate(stocks_list):
  plt.subplot(10,1,count+1)
  Q8_A(stock)
```
